```python
from crewai import LLM, Agent, Task, Crew

from dotenv import load_dotenv
load_dotenv()
import os
api_key= os.getenv("google2")
#os.environ["HUGGINGFACEHUB_API_KEY"]= os.getenv("HF_TOKEN")
llm= LLM(
      model="huggingface/openai/gpt-oss-20b",
    
)

```

```python
research_agent= Agent(
    role= "Teacher",
    goal= "Give a brief description about what is physics",
    backstory= "You are a great physics teacher and your student asked about the physics",
    llm= llm

)
    
```

```python
task= Task(
    agent= research_agent,
    description= "Find the revelant information about physics",
    expected_output= "5 bullets of very important information about physics",
    
)
```

```python
crew= Crew(
    verbose= True,
    agents= [research_agent],
    tasks= [task]
)
```

```python
result= crew.kickoff()
```

```python
result
```

```python
gk_agent= Agent(
    role= 'Intelligent Person',
    goal= 'Give a description about any topic in the world',
    backstory= 'You are one of the most intelligent person in the world and has high IQ. You have knowledge about almost anything in the world',
    llm= llm,

    max_retry_limit= 10
)
```

```python
gktask= Task(
    description= 'Summarize about how the world evolve',
    expected_output= 'Information about how the world evolved with different theories given by different peoples like bigbang etc.',
    agent= gk_agent,
    markdown= True,
    #output_file= 'info.md'

    
)
```

```python
crew1= Crew(
    agents= [gk_agent],
    tasks= [gktask]
)
```

```python
crew1.kickoff()
```

```python
analysis_agent= Agent(
    role = "Professional Analyst",
    goal= "Analyze the given chunk of text",
    backstory= "You are a professional analyzer. You are good at analyzing the document and extract the key point and compare them",
    llm= llm
)
```

```python
analysis_task= Task(
    description= "Analyze the markdown of the evolution of world and identify the key points",
    expected_output= 'Analyze the evolution of world',
    agent= analysis_agent,
    output_file= 'analysis.docx',
    context= [gktask]
)
```

```python
crew2= Crew(
    agents= [gk_agent, analysis_agent],
    tasks= [gktask, analysis_task],
    verbose= True,
    #process= Process.hierarchial
)
```

```python
result= crew2.kickoff()
```

```python
print(result)
```
