```python
from crewai import Agent, LLM
import os
from dotenv import load_dotenv
load_dotenv()
apikey= os.getenv("google")
```

```python
llm= LLM(
    model= "gemini/gemini-2.0-flash",
    api_key= apikey
   
)
```

```python
sum_agent= Agent(
    role= "A maths teacher that can do sum of any given numbers but cannot do any other operations except sum",
    goal= "Sum the user given numbers correctly but say I dont know when the given operation is rest than +",
    backstory= "You are a professional maths teacher who have 20+ experience in teaching on both college and school level. You can do addition of any complex math numbers and has no any record of doint it wrong but gives a lot of wrong answer when it comes to other operations than sum.",
    llm= llm,
    verbose= True,
    max_execution_time= 5,
    allow_delegation= False,
    cache= False,
    allow_code_execution= False,
    max_retry_limit= 3,
    respect_context_window= False,
    
)
```

```python
result= sum_agent.kickoff(messages= "What is 2-3?")
```

```python
print(result)
```

```python
result= sum_agent.kickoff("Can you give me answer of 2+3?")
result
```

```python
print(result.raw)
```

```python
#for structured output
from pydantic import BaseModel, Field
from typing import List, Union, Optional

class structured (BaseModel):
    Question: Union[str, List[str]]= Field("The questions which is asked by user. Return a string only if user asked only one question but returns list of string of questions if user asked multiple questions")

    Answer: Union[str, List[str]]= Field("The answers of questions. Return a string only if user asked only one question but returns list of string of answers correspomding to the indexed question in Question list if user asked multiple questions")
```

```python
subtract_agent= Agent(
    role= "A maths teacher that can do subtraction of any given numbers but cannot do any other operations except difference",
    goal= "Subtract the user given numbers correctly but say I dont know when the given operation is rest than -",
    backstory= "You are a professional maths teacher who have 20+ experience in teaching on both college and school level. You can do subtraction of any complex math numbers and has no any record of doint it wrong but gives a lot of wrong answer when it comes to other operations than subtraction.",
    llm= llm,
    verbose= False,
    max_execution_time= 5,
    allow_delegation= False,
    cache= False,
    allow_code_execution= False,
    max_retry_limit= 3,
    respect_context_window= False,
    
)
```

```python
result= subtract_agent.kickoff(messages= "3-4= ?, 5-4= ?", response_format= structured)
result
```

```python
print(result)
```

```python
print(result.pydantic.Question)
```

```python
maths_agent= Agent(
    role= "A maths teacher that can solve any type of mathematic operation",
    goal= "Solve the user given mathematic problem that may contain any type of operators",
    backstory= "You are a professional maths teacher who have 20+ experience in teaching on both college and school level. You are well known about the mathematics rule like BODMAS and solve any maths questions from simple to complex one.",
    llm= llm,
    verbose= True,
    max_execution_time= 7,
    allow_delegation= False,
    cache= False,
    allow_code_execution= False,
    max_retry_limit= 3,
    respect_context_window= False,
    inject_date= False
)
```

```python
result= maths_agent.kickoff(messages= "100/2, 5+4, 5*4-2, 3-5, [3+{2-9/5(2+1)}]", response_format= structured)
```
